```python
import pandas as pd
```

```python
df_raw = pd.read_csv("MovieReview-Sample.csv", header = None, names = ["Index", "Content", "Rating"])
df_raw = df_raw[["Content", "Rating"]]
content = df_raw.Content

df_raw.head()
```

```python
from nltk.corpus import opinion_lexicon
```

```python
bing_pos = set(opinion_lexicon.positive())
bing_neg = set(opinion_lexicon.negative())
```

```python
def count_pos_neg(data, positive_dict, negative_dict):
# count of positive and negative words that appeared in each message
# net count which is calculated by positive count subtracting negative count. 
    poscnt = []
    negcnt = []
    netcnt = []

    for nrow in range(0,len(data)):
        text = data[nrow]
        
        qa = 0
        qb = 0

        for word in positive_dict :
            if (word in text) :
                qa = qa + 1

        for word in negative_dict :
            if (word in text) :
                qb = qb + 1

        qc = qa - qb

        poscnt.append(qa)
        negcnt.append(qb)
        netcnt.append(qc)

    return (poscnt, negcnt, netcnt)
```

```python
df_raw["Positive Bing"], df_raw["Negative Bing"], df_raw["Net Bing"] = count_pos_neg(content, bing_pos, bing_neg)
df_raw["Predicted Rating Bing"] = 0
df_raw.loc[df_raw["Net Bing"] > 0, "Predicted Rating Bing"] = 1
df_raw[["Content", "Positive Bing", "Negative Bing", "Net Bing", "Predicted Rating Bing"]].head()
```

```python
from sklearn.metrics import classification_report
```

```python
#target_names = ["0", "1"]
class_bing = classification_report(df_raw["Rating"], df_raw["Predicted Rating Bing"], output_dict = True)
print(class_bing)
```

```python
precision_0_bing = class_bing.get("0").get("precision")
precision_1_bing = class_bing.get("1").get("precision")
precision_bing = (precision_0_bing + precision_1_bing)/2

recall_0_bing = class_bing.get("0").get("recall")
recall_1_bing = class_bing.get("1").get("recall")
recall_bing = (recall_0_bing + recall_1_bing)/2

f_0_bing = class_bing.get("0").get("f1-score")
f_1_bing = class_bing.get("1").get("f1-score")
f_bing = (f_0_bing + f_1_bing)/2
```

## LM Dict

```python
def read_local_dictionary(file):
    # create dictionary list
    words_dict = []
    with open(file, "r") as f: 
        for line in f:
            t = line.strip().lower()
            words_dict.append(t)
    return words_dict
```

```python
lm_pos = read_local_dictionary('positive-words-LM.txt')
lm_neg = read_local_dictionary('negative-words-LM.txt')
```

```python
df_raw["Positive LM"], df_raw["Negative LM"], df_raw["Net LM"] = count_pos_neg(content, lm_pos, lm_neg)
df_raw["Predicted Rating LM"] = 0
df_raw.loc[df_raw["Net LM"] > 0, "Predicted Rating LM"] = 1
df_raw[["Content", "Positive LM", "Negative LM", "Net LM", "Predicted Rating LM"]].head()
```

```python
class_lm = classification_report(df_raw["Rating"], df_raw["Predicted Rating LM"], output_dict = True)
print(class_lm)
```

```python
precision_0_lm = class_lm.get("0").get("precision")
precision_1_lm = class_lm.get("1").get("precision")
precision_lm = (precision_0_lm + precision_1_lm)/2

recall_0_lm = class_lm.get("0").get("recall")
recall_1_lm = class_lm.get("1").get("recall")
recall_lm = (recall_0_lm + recall_1_lm)/2

f_0_lm = class_lm.get("0").get("f1-score")
f_1_lm = class_lm.get("1").get("f1-score")
f_lm = (f_0_lm + f_1_lm)/2
```

## Vader

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
analyzer = SentimentIntensityAnalyzer()
scores = [analyzer.polarity_scores(sentence) for sentence in content]
```

```python
neg_s = [i["neg"] for i in scores]
neu_s = [i["neu"] for i in scores]
pos_s = [i["pos"] for i in scores]
compound_s = [i["compound"] for i in scores]
```

```python
df_raw["Vader Negative"], df_raw["Vader Neutral"], df_raw["Vader Positive"], df_raw["Vader Compound"] = neg_s, neu_s, pos_s, compound_s
df_raw["Predicted Rating Vader"] = 0
df_raw.loc[df_raw["Vader Compound"] > 0, "Predicted Rating Vader"] = 1
```

```python
df_raw[["Content", "Vader Negative", "Vader Neutral", "Vader Positive", "Vader Compound", "Predicted Rating Vader"]].head(5)
```

```python
class_vader = classification_report(df_raw["Rating"], df_raw["Predicted Rating Vader"], output_dict = True)
print(class_vader)
```

```python
precision_0_vader = class_vader.get("0").get("precision")
precision_1_vader = class_vader.get("1").get("precision")
precision_vader = (precision_0_vader + precision_1_vader)/2

recall_0_vader = class_vader.get("0").get("recall")
recall_1_vader = class_vader.get("1").get("recall")
recall_vader = (recall_0_vader + recall_1_vader)/2

f_0_vader = class_vader.get("0").get("f1-score")
f_1_vader = class_vader.get("1").get("f1-score")
f_vader = (f_0_vader + f_1_vader)/2
```
